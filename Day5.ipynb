###Title: Exploratory Data Analysis Project

Dataset: Indian Employee Salary Dataset

###Load and understand the dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Name": ["Aman","Riya","Karan","Sneha","Rahul","Pooja","Vikas","Neha","Arjun","Meera","Rohit","Anjali"],
    "Age": [22,25,29,31,26,24,28,30,27,23,32,29],
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female","Male","Female","Male","Female"],
    "Department": ["IT","HR","IT","Finance","IT","HR","Finance","IT","HR","Finance","IT","HR"],
    "Experience": [1,2,5,7,3,2,6,4,3,1,8,5],
    "Salary": [25000,28000,45000,52000,36000,27000,50000,42000,34000,26000,60000,39000]
}

df = pd.DataFrame(data)
df.head()
df.info()
```

###Clean the dataset thoroughly

```python
df = df.drop_duplicates()
df["Gender"] = df["Gender"].str.capitalize()
```

###Perform meaningful analysis

```python
df.groupby("Department")["Salary"].mean()
df.groupby("Gender")["Salary"].mean()
df["Salary"].describe()
```

Visualization 1 – Salary Distribution

```python
plt.hist(df["Salary"], bins=5)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

###2 – Gender vs Salary

```python
df.groupby("Gender")["Salary"].mean().plot(kind="bar")
plt.title("Gender vs Average Salary")
plt.ylabel("Salary")
plt.show()
```

###3 – Department vs Salary

```python
df.groupby("Department")["Salary"].mean().plot(kind="bar")
plt.title("Department vs Average Salary")
plt.ylabel("Salary")
plt.show()
```

###4 – Experience vs Salary

```python
plt.scatter(df["Experience"], df["Salary"])
plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

###5 – Age vs Salary

```python
plt.scatter(df["Age"], df["Salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

6 – Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.select_dtypes(include=['number']).corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
