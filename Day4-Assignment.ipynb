### Score distribution plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female","Male","Female"],
    "Math_Score": [78, 85, 69, 92, 88, 74, 81, 95, 67, 89],
    "Science_Score": [72, 90, 65, 94, 86, 70, 80, 96, 60, 88],
    "English_Score": [75, 88, 70, 90, 84, 73, 79, 93, 66, 87],
    "Attendance": [82, 91, 76, 95, 89, 80, 85, 97, 74, 90]
}

df = pd.DataFrame(data)

df["Average_Score"] = df[["Math_Score","Science_Score","English_Score"]].mean(axis=1)

plt.hist(df["Average_Score"], bins=5)
plt.title("Score Distribution")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```

###Gender vs performance chart

```python
df.groupby("Gender")["Average_Score"].mean().plot(kind="bar")
plt.title("Gender vs Average Performance")
plt.ylabel("Average Score")
plt.show()
```

###Correlation heatmap

```python
sns.heatmap(df[["Math_Score","Science_Score","English_Score","Attendance"]].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

###Attendance vs score trend chart

```python
plt.scatter(df["Attendance"], df["Average_Score"])
plt.title("Attendance vs Average Score")
plt.xlabel("Attendance")
plt.ylabel("Average Score")
plt.show()
```
