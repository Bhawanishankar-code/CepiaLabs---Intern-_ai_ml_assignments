Problem Statement

Q1 Generate 10,000 random house prices using NumPy and:
(A)Calculate average price ??
(B)Remove outliers ??
(C)Visualize price ??distribution

 Ans.Q1.(A) calculate average price ??

```python
import numpy as np

prices = np.abs(np.random.normal(4500000, 1500000, 10000))
avg_price = np.mean(prices)

print("Average price:", round(avg_price, 2))
```

Ans.Q1.(B) Remove outliers ??

```python
q1, q3 = np.percentile(prices, [25, 75])
iqr = q3 - q1

low = q1 - 1.5 * iqr
high = q3 + 1.5 * iqr

filtered = prices[(prices >= low) & (prices <= high)]

print("New average:", round(np.mean(filtered), 2))
print("Outliers removed:", len(prices) - len(filtered))
```

Ans.Q1.(C) Visualize price ??distribution

```python
import matplotlib.pyplot as plt

plt.hist(filtered, bins=40)
plt.title("House Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```
