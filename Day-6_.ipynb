####Dataset: Marketing Campaign Dataset

#####Create Marketing Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(5)

data = {
    "Customer_ID": range(1,101),
    "Spending": np.random.randint(500, 20000, 100),
    "Engagement_Score": np.random.randint(10, 100, 100),
    "Clicks": np.random.randint(1, 50, 100)
}

df = pd.DataFrame(data)
df.head()
```

#####Measures of Central Tendency

```python
mean_spend = df["Spending"].mean()
median_spend = df["Spending"].median()
mode_spend = df["Spending"].mode()[0]

print(mean_spend, median_spend, mode_spend)
```

#####Measures of Dispersion

```python
range_spend = df["Spending"].max() - df["Spending"].min()
variance_spend = df["Spending"].var()
std_spend = df["Spending"].std()

print(range_spend, variance_spend, std_spend)
```

#####Quartiles and IQR

```python
q1 = df["Spending"].quantile(0.25)
q3 = df["Spending"].quantile(0.75)
iqr = q3 - q1

print(q1, q3, iqr)
```

#####Average Spending per Customer

```python
print(df["Spending"].mean())
```

#####Top 10% Spenders

```python
top_spenders = df[df["Spending"] >= df["Spending"].quantile(0.9)]
top_spenders.head()
```

#####Skewness in Engagement

```python
print(df["Engagement_Score"].skew())
```

#####Spending Distribution Plot

```python
plt.hist(df["Spending"], bins=10)
plt.title("Customer Spending Distribution")
plt.show()
```

Insights

1.)Average customer spending is around middle range

2.)Top 10% customers spend much more than others

3.)Engagement data shows slight skewness

4.)Spending values are widely spread

5.)Customer behavior is not evenly distributed.
