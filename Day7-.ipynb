Day7-Assignment – Email Marketing Probability Analysis

##Dataset: Email Marketing Campaign

######Create Email Campaign Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(7)

data = {
    "User_ID": range(1,201),
    "Clicked": np.random.choice([0,1], 200, p=[0.6,0.4]),
}

df = pd.DataFrame(data)
df["Converted"] = np.where(df["Clicked"]==1, np.random.choice([0,1], 200, p=[0.7,0.3]), 0)

df.head()
```

###Probability that a user clicks an email

```python
p_click = df["Clicked"].mean()
print(p_click)
```

#####Probability that a user converts given they clicked

```python
p_convert_given_click = df[df["Clicked"]==1]["Converted"].mean()
print(p_convert_given_click)
```

#####Conditional Probability Table

```python
table = pd.crosstab(df["Clicked"], df["Converted"], normalize='index')
print(table)
```

#####Visualization of Click vs Conversion

```python
table.plot(kind="bar")
plt.title("Click vs Conversion Probability")
plt.ylabel("Probability")
plt.show()
```

Insights

l•)Around some users click on marketing emails


,2.)Only a part of clicked users convert.

3.)Conversion depends strongly on clicking.

4.)Most users who do not click never convert.
