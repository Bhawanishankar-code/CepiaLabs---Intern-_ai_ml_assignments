```python
#Program 3 #Sum of first N natural numbers using while loop
n = 10
sum = 0
i = 1
while i <= n:
    sum += i
    i += 1
print("Sum is:", sum)
```

```python
#Program 5 #Find factorial of a number using for loop
n = 5
fact = 1
for i in range(1, n + 1):
    fact *= i
print(f"Factorial of {n} is {fact}")
```

```python
#program 6 #Function to check even or odd
def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(check_even_odd(7))
print(check_even_odd(10))
```

```python
#Program 7 # Function to find maximum of three numbers
def max_of_three(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

print(max_of_three(3, 7, 5))
```

```python
#Program 8 #Function to reverse a string
def reverse_string(s):
    return s[::-1]

print(reverse_string("college"))
```

```python
#program 9 #Function to calculate area of rectangle
def area_rectangle(length, width):
    return length * width

print(area_rectangle(5, 3))
```

```python
#Program 10 #Function to check prime number
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(11))
print(is_prime(4))
```

```python
#program 13 #Find largest number in list
nums = [1, 2, 3, 4, 5]
largest = nums[0]
for n in nums:
    if n > largest:
        largest = n
print("Largest:", largest)
```

```python
#program 14 #Remove duplicates from list
list_with_duplicates = [1, 2, 2, 3, 4, 4, 5]
unique_list = []
for item in list_with_duplicates:
    if item not in unique_list:
        unique_list.append(item)
print("Unique list:", unique_list)
```

```python
#Program 20 #Student Database Dictionary
student = {
    "roll_no": "INT2026-9293",
    "name": "Bhawani Shankar Malvi",
    "age": 21,
    "branch": "BCA",
    "subjects": ["Python", "Maths", "DBMS", "OS"],
    "marks": {"Python": 90, "Maths": 78, "DBMS": 90, "OS": 82},
    "attendance": 88.5
}

print("=== STUDENT DETAILS ===")
print(f"Name: {student['name']}")
print(f"Roll No: {student['roll_no']}")
print(f"Age: {student['age']}")
print(f"Branch: {student['branch']}")

print("\n=== SUBJECTS ===")
for sub in student["subjects"]:
    print(sub)

print("\n=== MARKS ===")
for subject, score in student["marks"].items():
    print(f"{subject}: {score}")

total = student["marks"]["Python"] + student["marks"]["Maths"] + student["marks"]["DBMS"] + student["marks"]["OS"]
avg = total / 4

print(f"\nTotal Marks: {total}")
print(f"Average: {avg:.1f}%")

student["semester"] = 4
student["marks"]["AI"] = 88

print("\n=== UPDATED RECORD ===")
print(f"Semester: {student['semester']}")
print(f"AI Marks: {student['marks']['AI']}")
print(f"Attendance: {student['attendance']}%")
```
